## Ejercicio 3

Dada una lista de valores en la cual todos los valores están duplicados excepto uno. Hallar el número único.

````
Input: nums = [2,2,1]
Output: 1

Input: nums = [4,1,2,1,2]
Output: 4
````

`````python
def unico(lista):
    aux = {}
    for num in lista:
        if num in aux:
            aux[num] += 1
        else:
            aux[num] = 1
    for num, count in aux.items():
        if count == 1:
            return num
nums = [2,2,1]
nums2=[4,1,2,1,2]
print(unico(nums), unico(nums2))
`````

    * Usar los siguientes parámetros:
        * `s0`: `100` (para ambas series)
        * `serie_1`: ` r=0.09, sigma=0.22`
        * `serie_2`: ` r=0.055, sigma=0.28`
        * `num_steps=500`
1. Grafique ambas series en un mismo gráfico. Se evaluará la calidad y la interpretabilidad del gráfico.
1. Realice un análisis de estadísitica descriptiva para cada una de las series.
1. ¿Cuál es la probabilidad de que la `serie_1` sea menor a la `serie_2`?
1. ¿Cuál es la probabilidad de obtener un valor superior a 500 en la `serie_2`? 
1. ¿Cuál es la probabilidad de que la `serie_1` arroje un valor entre 200 y 500?
1. ¿Cuál es la probabilidad de que la `serie_2` se encuentre entre su media más/menos un cuarto de su desvío estándar?
1. Realice un análisis comparativo de las series. Sea exhaustivo en su desarrollo.

`````python
#### NO MODIFICAR DESDE ACÁ ####

rng = np.random.default_rng(789)

def generar_serie(s0, r, sigma, t=1, num_steps=100):
    
    serie = [s0]

    for i in range(num_steps):
        z = rng.normal(0, 1)
        s_t = serie[i] * np.e**((r - (sigma**2) / 2) * t + z * sigma * np.sqrt(t))
        serie.append(s_t)

    return serie

#### NO MODIFICAR HASTA ACÁ ####
`````

`````python
serie_1 = generar_serie(100,0.09,0.22)
serie_2 = generar_serie(100,0.55,0.28)
y = []
for i in range(1000):
    y.append(i)
`````

`````python
import matplotlib.pyplot as plt

plt.plot(serie_1, color='orange', label='serie 1')
plt.plot(serie_2, color='black', label='serie 2')
plt.title("Gráfico")
plt.xlabel("Eje X")
plt.ylabel("Eje Y") 

plt.legend()
`````
