### Librerias y Dataset

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_digits
```

### Explicacion de la libreria
La libreria muestra arrays que representan fotos, de numeros escritos a mano <br>
en la parte de target se muestra que numero se escribio por lo tanto es la varibles objetivo

```python
digits = load_digits()
```

```python
im = digits.data[9].reshape(8,8)
im
```

```python
plt.imshow(im, 'grey')
plt.show()
```

```python
print(digits.target[9])
```

### Train split test


```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(digits.data, digits.target,train_size=0.8,random_state=123) 
y_train.shape
```

```python
lg =  LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
y_pred = lg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, cmap='crest', linewidth=.5)
plt.ylabel('True value')
plt.xlabel('Predicted value')
plt.show()
```
