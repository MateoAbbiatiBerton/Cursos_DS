```python
df = df[['RM','PTRATIO','LSTAT','MEDV']]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop(['MEDV'], axis=1), df['MEDV'], test_size=0.2, random_state=123)
print(len(y_train), len(y_test))
```

```python
# Estandarizacin de los datos 
sc_x = StandardScaler()
sc_y = StandardScaler()
x_train_std = sc_x.fit_transform(x_train)
x_test_std = sc_x.fit_transform(x_test)
y_train_std = sc_y.fit_transform(y_train.values.reshape(-1,1))
```

```python
slr = LinearRegression()
slr.fit(x_train_std,y_train_std)
```

```python
slr.coef_.shape
b0 = slr.intercept_
b1 = slr.coef_[0,0]
b2 = slr.coef_[0,1]
b3 = slr.coef_[0,2]
print(f'{b0} + {b1} x X1+ {b2} x X2 + {b3} x X3')
```

```python
y_pred_std = slr.predict(x_test_std)
y_pred = sc_y.inverse_transform(y_pred_std)
y_pred
```

```python
# R^2
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
